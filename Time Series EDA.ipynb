```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla = pdr.get_data_yahoo('TSLA')
```

```python
df_tesla['High'].plot(figsize = (12,4))
```

```python
# xlimit and ylimit

df_tesla['High'].plot(xlim=['2020-01-01', '2021-09-01'],figsize = (12,4))
```

```python
df_tesla['High'].plot(xlim=['2020-01-01', '2022-09-01'], ylim= [0,400],figsize = (12,4))
```

```python
index = df_tesla.loc['2020-01-01': '2022-09-01'].index
```

```python
share_open = df_tesla.loc['2020-01-01': '2022-09-01']['Open']
share_close = df_tesla.loc['2020-01-01': '2022-09-01']['Close']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
figure, axis = plt.subplots()
figure.autofmt_xdate()
axis.plot(index, share_open)
#axis.plot(index, share_close)
```

## Rolling and Expanding

```python
df_tesla['High'].rolling(10).mean().head(15)
```

```python
df_tesla['Open Monthly Rolling'] = df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla[['Open', 'Open Monthly Rolling']].plot(figsize= (12,5))
```
